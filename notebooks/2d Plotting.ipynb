```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

```python
x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
```

```python
help(np.meshgrid)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy[100, :]
```

```python
r = np.sqrt(xx**2 + yy**2)
```

```python
r
```

```python
plt.contour(xx,yy,r, 35)
plt.colorbar()
fig = plt.gcf()
fig.set_figwidth(12)
fig.set_figheight(8)
```

```python
plt.contourf(xx,yy,r, 50)
plt.colorbar()
fig = plt.gcf()
fig.set_figwidth(12)
fig.set_figheight(8)
```

```python
r1 = r
r2 = np.sqrt((xx-2)**2 + (yy-1)**2)

V = 1/r1 - 1/r2

plt.contourf(xx, yy, V, 15)
#plt.xlim(-1,3)
#plt.ylim(-1, 2)
```
