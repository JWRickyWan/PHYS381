```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# load data from csv into dataframe
# and set time as index
df = pd.read_csv('0.0-0.0-9.371069245283019-0.0.csv')
df = df.set_index('time')

# create plot
fig, ax = plt.subplots(1)

# set size (horizontal: 12, vertical: 9)
fig.set_figwidth(12)
fig.set_figheight(9)

# plot kinetic energy
df['T'].plot(label="Kinetic")
# plot potential energy
df['V'].plot(ax=ax, label="Potential")
# plot total energy
df['E'].plot(ax=ax)

# add legend
ax.legend(loc="upper left", bbox_to_anchor=(1,1))

# add x limits
ax.set_xlim(70,75)
# add title
plt.title("Energy in a chaotic pendulum")

# add text at bottom
plt.text(72,-1,"source data: 0.0-0.0-9.371069245283019-0.0.csv")
# add axis labels
ax.set_ylabel("Energy (J)")
```
